```python
import pandas as pd
import os
import pickle as pk
import xlsxwriter
# import xlwt 
# from xlwt import Workbook
# from xlrd import open_workbook
# from xlutils.copy import copy
os.chdir("..")
```

```python
path = os.getcwd( )+"/Ground Truth and Results/"
num_districts = 5
districts = ["agra", "allahabad", "lucknow", "meerut", "varanasi"]
base_date = '2012-03'
```

```python
df = []
datemap = []
for i in range(num_districts):
    df.append(pd.read_excel(pd.ExcelFile(path+districts[i]+"_big.xlsx"), 'Event categories'))
    df[i] = df[i].T
    datemap.append([])
```

```python
def take(x,ind):
    global datemap
    if x[6] == "YES":
        val = x[7]
        val = val.split(',')
        for i in range(len(val)):
            val[i] = val[i].strip()
        temp = [0, 0, 0, 0, 0]
        for i in range(4):
            if x[i+1] == "YES":
                temp[i] = 1
        if x[5] == "Local":
            temp[4] = 0
        elif x[5] == "State":
            temp[4] = 1
        else:
            temp[4] = 2
        
        datemap[ind].append([val,temp]);
    return

def greater_than(a,b):
    a = a.split('-')
    b = b.split('-')
    v1 = int(a[0])
    v2 = int(a[1])
    w1 = int(b[0])
    w2 = int(b[1])
    if v1 > w1:
        return 1
    if v1==w1 and v2 > w2:
        return 1
    return 0

def makemap(df, ind):
    temp = df[ind].apply(lambda x: take(x,ind))

for i in range(num_districts):
    makemap(df,i)
```

```python
result_sat1 = []
result_mm1 = []
file1 = open(os.getcwd( )+"/Ground Truth and Results/anomalies20.pk",'rb')
file2 = open(os.getcwd( )+"/Ground Truth and Results/anomalies25mm.pk",'rb')
for i in range(num_districts):
    result_sat1.append(pk.load(file1))
    result_mm1.append(pk.load(file2))
file1.close()
file2.close()

result_sat = []
result_mm = []

for i in range(num_districts):
    temp1 = []
    temp2 = []
    for rr in result_sat1[i]:
        if greater_than(rr, base_date):
            temp1.append(rr)
    for rr in result_mm1[i]:
        if greater_than(rr, base_date):
            temp2.append(rr)
    result_sat.append(temp1)
    result_mm.append(temp2)
```

```python
workbook = xlsxwriter.Workbook(os.getcwd( )+"/Ground Truth and Results/result.xlsx")
bold = workbook.add_format({'bold': 1})

category = ["Political","Social","Crime","Calamity"]
level = ["Local", "State", "National"]
num_level = len(level)
num_cat = len(category)

num_anomalies_sat = 20
num_anomalies_mm = 20

for i in range(num_districts):
    result_sat[i] = result_sat[i][0:num_anomalies_sat]
    result_mm[i] = result_mm[i][0:num_anomalies_mm]

recall_avg_sat = [[[[0.0 for tt in range(num_level)] for kk in range(num_cat)] for j in range(num_anomalies_sat)] for jj in range(2)]
recall_overall_avg_sat = [[[0.0 for tt in range(num_level)] for j in range(num_anomalies_sat)] for jj in range(2)]
precision_avg_sat = [[0.0 for j in range(num_anomalies_sat)] for jj in range(2)]
recall_avg_mm = [[[[0.0 for tt in range(num_level)] for kk in range(num_cat)] for j in range(num_anomalies_mm)] for jj in range(2)]
recall_overall_avg_mm = [[[0.0 for tt in range(num_level)] for j in range(num_anomalies_mm)] for jj in range(2)]
precision_avg_mm = [[0.0 for j in range(num_anomalies_mm)] for jj in range(2)]

for i in range(num_districts):
#     sheet = wy.get_sheet(i)
#     sheet = wb.add_sheet(districts[i])
    sheet = workbook.add_worksheet(districts[i])
    drift = len(result_sat[i])+3
    for u in range(len(result_sat[i])):
        sheet.write_number(u+2, 0, (u+1), bold)
    for u in range(len(result_mm[i])):
        sheet.write_number(u+drift+2, 0, (u+1), bold)
    category.append("Overall")
    for cat in range(len(category)):
        sheet.write_string(0, len(level)*cat+1, category[cat], bold)
        sheet.write_string(drift, len(level)*cat+1, category[cat], bold)
        for ll in range(len(level)):
            sheet.write_string(1, len(level)*cat+1+ll, level[ll], bold)
            sheet.write_string(1+drift, len(level)*cat+1+ll, level[ll], bold)
    
    sheet.write_string(0, len(level)*len(category)+1, "Precision", bold)
    sheet.write_string(drift, len(level)*len(category)+1, "Precision", bold)
    category.pop(len(category)-1)
    
    arr = [[0.0 for i in range(num_level)] for j in range(num_cat)]
    arr_overall = [0.0 for i in range(num_level)]
    for x in datemap[i]:
        arr_overall[x[1][4]] += 1
        for j in range(num_cat):
            if x[1][j] == 1:
                arr[j][x[1][4]] += 1
    recall = [[[0 for tt in range(len(datemap[i]))] for kk in range(num_level)] for j in range(num_cat)]
    recall_overall = [[0 for tt in range(len(datemap[i]))] for kk in range(num_level)]
    
    
    chart1 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})
    chart2 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})
    
    cc = 0
    acc = 0.0
    for rr in result_sat[i]:
        xxx = 0
        found = 0
        for x in datemap[i]:
            if rr in x[0]:
                found = 1
                for j in range(num_cat):
                    if x[1][j] == 1:
                        recall[j][x[1][4]][xxx] = 1
                        recall_overall[x[1][4]][xxx] = 1
            xxx += 1
        acc += found
        for j in range(num_cat):
            for x in range(num_level):
                numx = 0.0
                for y in range(len(datemap[i])):
                    if recall[j][x][y] == 1:
                        numx += 1.0
                if arr[j][x] != 0:
                    sheet.write_number(2+cc, num_level*j+1+x, numx/arr[j][x])
                    recall_avg_sat[0][cc][j][x] += numx/arr[j][x]
                    recall_avg_sat[1][cc][j][x] += 1
                else:
                    sheet.write_number(2+cc, num_level*j+1+x, -1)
                if cc == len(result_sat[i])-1:
                    if arr[j][x] != 0:
                        chart1.add_series({
                            'name':       category[j]+' '+level[x],
                            'categories': [districts[i], 2, 0, 2+len(result_sat[i])-1, 0],
                            'values':     [districts[i], 2, num_level*j+1+x, 2+len(result_sat[i])-1, num_level*j+1+x],
                        })
            
        for x in range(num_level):
            numx = 0.0
            for y in range(len(datemap[i])):
                if recall_overall[x][y] == 1:
                    numx += 1.0
            if arr_overall[x] != 0:
                sheet.write_number(2+cc, num_level*num_cat+1+x, numx/arr_overall[x])
                recall_overall_avg_sat[0][cc][x] += numx/arr_overall[x]
                recall_overall_avg_sat[1][cc][x] += 1
            else:
                sheet.write_number(2+cc, num_level*num_cat+1+x, -1)
            if cc == len(result_sat[i])-1:
                if arr_overall[x] != 0:
                    chart2.add_series({
                        'name':       level[x],
                        'categories': [districts[i], 2, 0, 2+len(result_sat[i])-1, 0],
                        'values':     [districts[i], 2, num_level*num_cat+1+x, 2+len(result_sat[i])-1, num_level*num_cat+1+x],
                    })
        
        sheet.write_number(2+cc, (num_level+1)*num_cat, acc/(cc+1.0))
        precision_avg_sat[0][cc] += acc/(cc+1.0)
        precision_avg_sat[1][cc] += 1
        if cc == len(result_sat[i])-1:
            chart2.add_series({
                'name':       'Precision',
                'categories': [districts[i], 2, 0, 2+len(result_sat[i])-1, 0],
                'values':     [districts[i], 2, (num_level+1)*num_cat, 2+len(result_sat[i])-1, (num_level+1)*num_cat],
            })
        cc += 1
    
    chart1.set_title ({'name': 'Satellite'})
    chart1.set_x_axis({'name': 'Number of anomalies'})
    chart1.set_y_axis({'name': 'Recall'})
    chart1.set_style(15)
    chart1.set_size({'width': 1100, 'height': 500})
    sheet.insert_chart('S1', chart1)
    
    chart2.set_title ({'name': 'Satellite'})
    chart2.set_x_axis({'name': 'Number of anomalies'})
    chart2.set_style(15)
    chart2.set_size({'width': 1100, 'height': 500})
    sheet.insert_chart('AK1', chart2)
        
    chart1 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})
    chart2 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})
    
    recall = [[[0 for tt in range(len(datemap[i]))] for kk in range(num_level)] for j in range(num_cat)]
    recall_overall = [[0 for tt in range(len(datemap[i]))] for kk in range(num_level)]
    cc = 0
    acc = 0.0
    for rr in result_mm[i]:
        xxx = 0
        found = 0
        for x in datemap[i]:
            if rr in x[0]:
                found = 1
                for j in range(num_cat):
                    if x[1][j] == 1:
                        recall[j][x[1][4]][xxx] = 1
                        recall_overall[x[1][4]][xxx] = 1
            xxx += 1
        acc += found
        for j in range(num_cat):
            for x in range(num_level):
                numx = 0.0
                for y in range(len(datemap[i])):
                    if recall[j][x][y] == 1:
                        numx += 1.0
                if arr[j][x] != 0:
                    sheet.write_number(2+cc+drift, num_level*j+1+x, numx/arr[j][x])
                    recall_avg_mm[0][cc][j][x] += numx/arr[j][x]
                    recall_avg_mm[1][cc][j][x] += 1
                else:
                    sheet.write_number(2+cc+drift, num_level*j+1+x, -1)
                if cc == len(result_mm[i])-1:
                        if arr[j][x] != 0:
                            chart1.add_series({
                                'name':       category[j]+' '+level[x],
                                'categories': [districts[i], 2+drift, 0, 2+len(result_mm[i])-1+drift, 0],
                                'values':     [districts[i], 2+drift, num_level*j+1+x, 2+len(result_mm[i])-1+drift, num_level*j+1+x],
                            })
                    
        for x in range(num_level):
            numx = 0.0
            for y in range(len(datemap[i])):
                if recall_overall[x][y] == 1:
                    numx += 1.0
            if arr_overall[x] != 0:
                sheet.write_number(2+cc+drift, num_level*num_cat+1+x, numx/arr_overall[x])
                recall_overall_avg_mm[0][cc][x] += numx/arr_overall[x]
                recall_overall_avg_mm[1][cc][x] += 1
            else:
                sheet.write_number(2+cc+drift, num_level*num_cat+1+x, -1)
            if cc == len(result_mm[i])-1:
                if arr_overall[x] != 0:
                    chart2.add_series({
                        'name':       level[x],
                        'categories': [districts[i], 2+drift, 0, 2+len(result_mm[i])-1+drift, 0],
                        'values':     [districts[i], 2+drift, num_level*num_cat+1+x, 2+len(result_mm[i])-1+drift, num_level*num_cat+1+x],
                    })
        if cc == len(result_mm[i])-1:
            chart2.add_series({
                'name':       'Precision',
                'categories': [districts[i], 2+drift, 0, 2+len(result_mm[i])-1+drift, 0],
                'values':     [districts[i], 2+drift, (num_level+1)*num_cat, 2+len(result_mm[i])-1+drift, (num_level+1)*num_cat],
            })
        sheet.write_number(2+cc+drift, (num_level+1)*num_cat, acc/(cc+1.0))
        precision_avg_mm[0][cc] += acc/(cc+1.0)
        precision_avg_mm[1][cc] += 1
        cc += 1
    
    chart1.set_title ({'name': 'Mass Media'})
    chart1.set_x_axis({'name': 'Number of anomalies'})
    chart1.set_y_axis({'name': 'Recall'})
    chart1.set_style(15)
    chart1.set_size({'width': 1100, 'height': 500})
    sheet.insert_chart('S30', chart1)
    
    chart2.set_title ({'name': 'Mass Media'})
    chart2.set_x_axis({'name': 'Number of anomalies'})
    chart2.set_style(15)
    chart2.set_size({'width': 1100, 'height': 500})
    sheet.insert_chart('AK30', chart2)
    
########################## Average ###############################################
    
sheet = workbook.add_worksheet("Average")
drift = num_anomalies_sat+3
for u in range(num_anomalies_sat):
    sheet.write_number(u+2, 0, (u+1), bold)
for u in range(len(result_mm[i])):
    sheet.write_number(u+drift+2, 0, (u+1), bold)
category.append("Overall")
for cat in range(len(category)):
    sheet.write_string(0, len(level)*cat+1, category[cat], bold)
    sheet.write_string(drift, len(level)*cat+1, category[cat], bold)
    for ll in range(len(level)):
        sheet.write_string(1, len(level)*cat+1+ll, level[ll], bold)
        sheet.write_string(1+drift, len(level)*cat+1+ll, level[ll], bold)

sheet.write_string(0, len(level)*len(category)+1, "Precision", bold)
sheet.write_string(drift, len(level)*len(category)+1, "Precision", bold)
category.pop(len(category)-1)

chart1 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})
chart2 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})

for cc in range(num_anomalies_sat):
    for j in range(num_cat):
        for x in range(num_level):
            if recall_avg_sat[1][cc][j][x] != 0:
                sheet.write_number(2+cc, num_level*j+1+x, recall_avg_sat[0][cc][j][x]/recall_avg_sat[1][cc][j][x])
            else:
                sheet.write_number(2+cc, num_level*j+1+x, -1)
            if cc == num_anomalies_sat-1:
                if recall_avg_sat[1][cc][j][x] != 0:
                    chart1.add_series({
                        'name':       category[j]+' '+level[x],
                        'categories': ["Average", 2, 0, 2+num_anomalies_sat-1, 0],
                        'values':     ["Average", 2, num_level*j+1+x, 2+num_anomalies_sat-1, num_level*j+1+x],
                    })

    for x in range(num_level):
        if recall_overall_avg_sat[1][cc][x] != 0:
            sheet.write_number(2+cc, num_level*num_cat+1+x, recall_overall_avg_sat[0][cc][x]/recall_overall_avg_sat[1][cc][x])
        else:
            sheet.write_number(2+cc, num_level*num_cat+1+x, -1)
        if cc == num_anomalies_sat-1:
            if recall_overall_avg_sat[1][cc][x] != 0:
                chart2.add_series({
                    'name':       level[x],
                    'categories': ["Average", 2, 0, 2+num_anomalies_sat-1, 0],
                    'values':     ["Average", 2, num_level*num_cat+1+x, 2+num_anomalies_sat-1, num_level*num_cat+1+x],
                })

    sheet.write_number(2+cc, (num_level+1)*num_cat, precision_avg_sat[0][cc]/precision_avg_sat[1][cc])
    if cc == num_anomalies_sat-1:
        chart2.add_series({
            'name':       'Precision',
            'categories': ["Average", 2, 0, 2+num_anomalies_sat-1, 0],
            'values':     ["Average", 2, (num_level+1)*num_cat, 2+num_anomalies_sat-1, (num_level+1)*num_cat],
        })
    cc += 1

chart1.set_title ({'name': 'Satellite'})
chart1.set_x_axis({'name': 'Number of anomalies'})
chart1.set_y_axis({'name': 'Recall'})
chart1.set_style(15)
chart1.set_size({'width': 1100, 'height': 500})
sheet.insert_chart('S1', chart1)

chart2.set_title ({'name': 'Satellite'})
chart2.set_x_axis({'name': 'Number of anomalies'})
chart2.set_style(15)
chart2.set_size({'width': 1100, 'height': 500})
sheet.insert_chart('AK1', chart2)

chart1 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})
chart2 = workbook.add_chart({'type': 'scatter','subtype': 'straight_with_markers'})

for cc in range(num_anomalies_mm):
    for j in range(num_cat):
        for x in range(num_level):
            if recall_avg_mm[1][cc][j][x] != 0:
                sheet.write_number(2+cc+drift, num_level*j+1+x, recall_avg_mm[0][cc][j][x]/recall_avg_mm[1][cc][j][x])
            else:
                sheet.write_number(2+cc+drift, num_level*j+1+x, -1)
            if cc == num_anomalies_mm-1:
                    if recall_avg_mm[1][cc][j][x] != 0:
                        chart1.add_series({
                            'name':       category[j]+' '+level[x],
                            'categories': ["Average", 2+drift, 0, 2+num_anomalies_mm-1+drift, 0],
                            'values':     ["Average", 2+drift, num_level*j+1+x, 2+num_anomalies_mm-1+drift, num_level*j+1+x],
                        })

    for x in range(num_level):
        if recall_overall_avg_mm[1][cc][x] != 0:
            sheet.write_number(2+cc+drift, num_level*num_cat+1+x, recall_overall_avg_mm[0][cc][x]/recall_overall_avg_mm[1][cc][x])
        else:
            sheet.write_number(2+cc+drift, num_level*num_cat+1+x, -1)
        if cc == num_anomalies_mm-1:
            if recall_overall_avg_mm[1][cc][x] != 0:
                chart2.add_series({
                    'name':       level[x],
                    'categories': ["Average", 2+drift, 0, 2+num_anomalies_mm-1+drift, 0],
                    'values':     ["Average", 2+drift, num_level*num_cat+1+x, 2+num_anomalies_mm-1+drift, num_level*num_cat+1+x],
                })
    if cc == num_anomalies_mm-1:
        chart2.add_series({
            'name':       'Precision',
            'categories': ["Average", 2+drift, 0, 2+num_anomalies_mm-1+drift, 0],
            'values':     ["Average", 2+drift, (num_level+1)*num_cat, 2+num_anomalies_mm-1+drift, (num_level+1)*num_cat],
        })
    sheet.write_number(2+cc+drift, (num_level+1)*num_cat, precision_avg_mm[0][cc]/precision_avg_mm[1][cc])
    cc += 1

chart1.set_title ({'name': 'Mass Media'})
chart1.set_x_axis({'name': 'Number of anomalies'})
chart1.set_y_axis({'name': 'Recall'})
chart1.set_style(15)
chart1.set_size({'width': 1100, 'height': 500})
sheet.insert_chart('S30', chart1)

chart2.set_title ({'name': 'Mass Media'})
chart2.set_x_axis({'name': 'Number of anomalies'})
chart2.set_style(15)
chart2.set_size({'width': 1100, 'height': 500})
sheet.insert_chart('AK30', chart2)

# wb.save(os.getcwd( )+"/Ground Truth and Results/graphs.xls")
workbook.close()
```
