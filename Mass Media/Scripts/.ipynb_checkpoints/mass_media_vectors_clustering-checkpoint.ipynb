## Data Input

```python
directory_in_str = "/Users/abhishekyadav/Documents/Study/BTP/city_to_article/Data/"
state = "up"
directory = directory_in_str + state
result_dir = directory_in_str + state + "_result"
if not os.path.exists(result_dir):
    print("Error: " + result_dir + " Directory doesn't exist")
```

```python
temp = []
tempvocablory = set()
for filename in os.listdir(directory):
    if not(filename.endswith("big.pk")):
        continue
    file1 = open(result_dir + "/" + filename,'rb')
    data = pk.load(file1)
    vocablury = pk.load(file1)
    file1.close()
    
    for key in vocablury.keys():
        tempvocablory.add(key)
    temp.append([data, vocablury])

i=0
vocabulary = {}
inttokeyword = {}
for x in sorted(tempvocablory):
    vocabulary[x] = i
    inttokeyword[i] = x
    i+=1

tempx = []
docs = {}
for x in temp:
    for z in x[0]:
        xx = np.zeros((len(vocabulary)))
        ss = ""
        for y in x[1].keys():
            xx[vocabulary[y]] = x[0][z][int(x[1][y])]
            if(xx[vocabulary[y]] != 0):
                ss = ss + " " + y
        tempx.append(xx)
        if z in docs:
            docs[z].append(ss)
        else:
            docs[z] = [ss]

X = np.array(tempx)
```

## Clustering directly

```python
clusters = 3
kmeans = KMeans(n_clusters=clusters, random_state=0).fit(X)
labels = kmeans.labels_
labels
```

```python
keywords = [{} for i in range(clusters)]
for i in range(len(labels)):
    for j in range(len(X[i])):
        if(X[i][j]!=0):
            if(inttokeyword[j] in keywords[labels[i]]):
                keywords[labels[i]][inttokeyword[j]] += X[i][j]
            else:
                keywords[labels[i]][inttokeyword[j]] = X[i][j]
```

```python
final_keywords = []
for x in keywords:
    tt = sorted(x.items(), key=lambda kv: kv[1], reverse=True)
    final_keywords.append([tt[i][0] for i in range(len(tt))])    
```

```python
for x in final_keywords:
    print(x[:10])
```

## Global vs Local anomalies

```python
anomalies = {}
for x in temp:
    for z in x[0]:
        if z in anomalies:
            anomalies[z] += 1
        else:
            anomalies[z] = 1
anomalies
```

```python
def tfidf(arr, relevant_words):
    index_of_words={}
    i=0
    for j in relevant_words:
        index_of_words[j]=i
        i+=1
    
    sz = len(relevant_words)
    doc_count = len(arr)
    tf = [ [0.0 for j in range(doc_count)] for i in range(sz)]
    idf = [ 0.0 for i in range(sz)]
    tf_idf = [ [0.0 for j in range(doc_count)] for i in range(sz)]
    av_tfidf = [ 0.0 for i in range(sz)]
    total_occurence = [ 0.0 for i in range(sz)]

    for i in range(len(arr)):
        s=arr[i].split()
        doc_occurence=[ 0.0 for i in range(sz)]
        for w in s:
            if w==".":
                continue
            if w in relevant_words:
                idx=index_of_words[w]
                tf[idx][i]+=1
                if doc_occurence[idx]==0:
                    doc_occurence[idx]=1
                    total_occurence[idx]+=1
        for j in range(sz):
            tf[j][i]/=len(s)
    for i in range(sz):
        idf[i]=math.log(doc_count/(1+total_occurence[i]))
    for i in range(doc_count):
        for j in range(sz):
            tf_idf[j][i]=tf[j][i]*idf[j]
            av_tfidf[j]+=tf_idf[j][i]
    final_words=[]
    i=0
    for j in relevant_words:
    #     final_words.append(((av_tfidf[i]**5)*map_to_rake[j],j))
        final_words.append(((av_tfidf[i]),j))
        i+=1
    final_words.sort(key=lambda x:x[0])
    final_words.reverse()
    return final_words

global_ano = []
local_ano = []
for x in anomalies:
    if anomalies[x] > 2:
        global_ano.append(docs[x])
    elif anomalies[x] == 1:
        local_ano.append(docs[x])
```

### Global

```python
global_ano1 = list(chain.from_iterable(global_ano))

relevant_words = set()
for key in global_ano1:
    word = key.split()
    for w in word:
        relevant_words.add(w)

result = tfidf(global_ano1, relevant_words)
result[:10]
```

### Local

```python
local_ano1 = list(chain.from_iterable(local_ano))
relevant_words = set()
for key in local_ano1:
    word = key.split()
    for w in word:
        relevant_words.add(w)
result = tfidf(local_ano1, relevant_words)
result[:10]
```

```python
client = MongoClient('mongodb://10.237.27.104:27017/')
art_db = client['media-db']
temp = pd.Series.from_csv("/Users/abhishekyadav/Documents/Study/BTP/city_to_article/SmartStoplist.txt",sep=' ',index_col=1)
temp = temp.tolist()[1:]
rake_object = rake.Rake(temp,[],max_length=4)  # min no. of chars=3, max no. of words=5, min frequency=1
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
arr = []
for filename in os.listdir(directory):
    arr = []
    if not(filename.endswith("big.pk")):
        continue
    file1 = open(result_dir + "/" + filename,'rb')
    data = pk.load(file1)
    voc = pk.load(file1)
    file1.close()
    
    myarr = ["" for x in range(len(voc))]
    for x in voc:
        myarr[voc[x]] = x
    
    newdata = {}
    for x in data:
        for y in range(len(data[x])):
            if data[x][y] != 0.0:
                if x not in newdata:
                    newdata[x] = set()
                newdata[x].add(myarr[y])
    
    query = filename.split("_")[1];
    print(query)
    map1 = (state_map if queryf==1 else city_map)
    
    if query in map1.keys():
        for it in map1[query]:
            cursor = art_db.articles.find({'_id': it})
            for art in cursor:
                tempdate = art['publishedDate'][:7]
                if tempdate in newdata:
                    text = art['text']
                    sentences = sent_tokenize(text)
                    localset = set()
                    linearr = ''
                    for sent in sentences:
                        line = ''
                        for word in word_tokenize(sent):
                            tword = wordnet_lemmatizer.lemmatize(word)
                            if len(tword) > 2:
                                tword = tword.lower()
                                localset.add(tword)
                                line += ' ' + tword
                        linearr += line + '.'
                    
                    cc = 0
                    for x in newdata[tempdate]:
                        if x in localset:
                            cc+=1
                            break
                    
                    if cc > 0:
                        arr.append(linearr)

    break

arr
```
